# Mission 3

Create a function that takes a dataframe as an argument and print the filling rate (the percentage of non-empty rows) of each of its columns. 

For example :

*   column 1 : 73%
*   column 2 : 99%


```
def check_filling_rate(dataframe):
  check = dataframe.notna().mean()
  column_number = 0
  for col_name in df.columns:
    print (f"{col_name} : {(round(check[column_number], 2)*100).astype(int)}%")
    column_number +=1

check_filling_rate(df)
```

# Mission 4

How many columns belong to each quartile of the filling rate? In other words, how many columns are filled at 0%, between 0% and 25%, between 25% and 50%...


```
check = df.notna().mean()
check2 = pd.DataFrame(check)
check2.reset_index(inplace=True)

check_zero = check2.loc[check2[0] == 0]
print (f"{len(check_zero.index)} columns are filled at 0%")

check_first_quartile = check2.loc[(check2[0] > 0) & (check2[0] <=0.25)]
print (f"{len(check_first_quartile.index)} columns are filled between 0% and 25%")

check_second_quartile = check2.loc[(check2[0] > 0.25) & (check2[0] <=0.5)]
print (f"{len(check_second_quartile.index)} columns are filled between 25% and 50%")

check_third_quartile = check2.loc[(check2[0] > 0.5) & (check2[0] <=0.75)]
print (f"{len(check_third_quartile.index)} columns are filled between 50% and 75%")

check_fourth_quartile = check2.loc[(check2[0] > 0.75) & (check2[0] <=1)]
print (f"{len(check_fourth_quartile.index)} columns are filled between 75% and 100%")
```

# Mission 6

A dataset with too few data is not very interesting. After some analysis we realized that we can delete columns with more than 60% null values, except the columns pnns_groups_1, pnns_groups_2, nutrition-score-fr_100g and nutrition-score-uk_100g. Indeed these columns are going to be useful later.

If it's not done already, delete also the following columns since they won't help to solve our problem:
* creator
* created_t
* created_datetime
* last_modified_t
* brands_tags
* countries
* countries_tags
* additives_tags
* states
* states_tags

```
df2 = df_france.copy()
df_keep_columns = df2[["pnns_groups_1", "pnns_groups_2", "nutrition-score-fr_100g", "nutrition-score-uk_100g"]]
df2.drop(columns = ["creator", "created_t", "created_datetime", "last_modified_t", "brands_tags", "countries", "countries_tags", "additives_tags", "states", "states_tags", "pnns_groups_1", "pnns_groups_2", "nutrition-score-fr_100g", "nutrition-score-uk_100g"], inplace=True)


perc = 40.0 # Like N %
min_count =  int(((100-perc)/100)*df2.shape[0] + 1)
print (min_count)
df2 = df2.dropna( axis=1, thresh=min_count)

df_final = pd.merge(df2, df_keep_columns, how='inner', left_index=True, right_index=True)
print (df_final.info())
```

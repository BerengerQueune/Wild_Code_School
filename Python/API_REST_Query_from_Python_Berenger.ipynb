```python
import requests
import pandas as pd
```

```python
link = 'https://data.nantesmetropole.fr/api/records/1.0/search/?dataset=244400404_stations-velos-libre-service-nantes-metropole-disponibilites&q=&facet=banking&facet=bonus&facet=status&facet=contract_name'
r = requests.get(link)
pd.json_normalize(r.json())
```

```python
my_data_frame = pd.json_normalize(r.json(),  record_path= ["records"])[["fields.address", "fields.available_bikes", "fields.bike_stands"]]

my_data_frame
```

```python
my_data_frame["taux_remplissage"] = my_data_frame["fields.available_bikes"] / my_data_frame["fields.bike_stands"]
my_data_frame.rename(columns={'fields.address': 'adresse'}, inplace=True)
my_data_frame.drop(['fields.available_bikes', 'fields.bike_stands'], axis=1, inplace=True)
my_data_frame
```
